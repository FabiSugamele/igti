```
# digitar o item compra e quantidade
# verificar o valor total
# quando encerrar a compra informar o valor da compra

produtos = ["Hamburguer", "Batata-frita", "Combo", "Refrigerante"]
precos = [15,10.50,20,5]
itens_comprados = []
valor_compra = 0

item_compra = int (input("Qual o identificador do item da compra? "))
itens_comprados.append(produtos[item_compra])
informacao = input("Digite a quantidade do item: ")

while ( not informacao.isdigit()):
    informacao = input("A quantidade digitada dev ser um número. Digite novamente")
quantidade = int (informacao)

while item_compra != -1:
  valor_compra = valor_compra + (precos[item_compra] * quantidade)
  item_compra = int (input("Qual o identificador do item da compra? "))
  if (item_compra==-1):
    break
  itens_comprados.append(produtos[item_compra])
  informacao = input("Qual a quantidade consumida: ")
  while (not informacao.isdigit()):
    informacao = input("A quantidade digitada dev ser um número. Digite novamente")
  quantidade = int (informacao)

print("***Relátorio final da compra***")
for item in itens_comprados:
  print("Comprado: {}".format(item))
print("*******************************")
print("Total da compra R$ {}".format(valor_compra))  
```
