```
# jogo de dados
# Escolho o valor entre 1 e 6
# Escolho quanto quero apostar, valor máximo de 100(valor que tenho no bolso)
# Jogo de 2 dados
# Se 1 dado igual, ganho 2 vezes o valor apostado
# Se 2 dados iguais, ganho 10 vezes o valor apostado
# Se não não der nenhum dado igual, perco meu dinheiro

from random import randrange
bolso = 100
numero_apostado = int(input("Qual numero você aposta? "))
valor_aposta = float(input("Qual valor você aposta? "))
if valor_aposta > bolso:
  print("Você não tem esse dinheiro")
else:
  bolso = bolso - valor_aposta
  dado1 = randrange(1,6)
  dado2 = randrange(1,6)
  mensagem = "Você ganhou {} e está com R$ {}"
  print ("Sorteados os dados {} e {}".format(dado1, dado2))
  resultado = 0
  if (dado1==numero_apostado) and (dado2==numero_apostado):
    resultado = valor_aposta * 10
    bolso = resultado + bolso
    print(mensagem.format(resultado,bolso))
  elif (dado1==numero_apostado) or (dado2==numero_apostado):
    resultado = valor_aposta * 2
    bolso = resultado + bolso
    print(mensagem.format(resultado,bolso))
  else:
    print("Você errou, agora tem no bolso R$ {}".format(bolso))
```

```
# Calculo do IMC
# Formula do IMC = peso / (altura**2)
# abaixo de 17 => muito abaixo do peso
# entre 17 e 18,5 => abaixo do peso
# entre 18,5 e 25 => peso normal
# entre 25 e 30 => peso acima do normal
# entre 30 e 35 => obesidade grau I
# entre 35 e 40 => obesidade grau II
# acima de 40 => obesidade morbida

print("Olá, vamos descobrir qual o seu IMC.")
peso = float (input("Digite seu peso: "))
altura = float(input("Digite sua altura: "))
imc = peso / (altura**2)
print("Seu IMC é {}".format(imc))
if imc < 17:
  print("Você está muito abaixo do peso")
elif (imc >= 17) and (imc <= 18.5):
  print("Você está abaixo do peso")
elif (imc >= 18.5) and (imc <= 25):
  print("Seu peso está normal")
elif (imc >= 25) and (imc <= 30):
  print("Você está acima do peso")
elif (imc >= 30) and (imc <= 35):
  print("Você está com obesidade grau I")
elif (imc >= 35) and (imc <= 40):
  print("Você está com obesidade grau I")
elif imc >= 40:
  print("Você está com obesidade morbida")
```
